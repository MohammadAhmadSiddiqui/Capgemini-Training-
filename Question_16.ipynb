Scenario:

We want to classify whether a restaurant order is “Large” (1) or “Small” (0) based on the number
  of items ordered.
 - If items ≥ 3 → Large order
 - If items < 3 → Small orde

```python
import numpy as np

# Step 1: Training Data
# X = number of items ordered
# y = order size (0 = Small, 1 = Large)

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([0, 0, 1, 1, 1])   # <3 small, >=3 large

# Step 2: Initialize weights and bias
weights = np.random.rand(1)
bias = np.random.rand(1)
learning_rate = 0.1

# Step 3: Activation Function (Step function)
def activation(z):
    if z >= 0:
        return 1
    else:
        return 0

# Step 4: Training Loop
for epoch in range(10):
    print(f"Epoch {epoch+1}")

    for i in range(len(X)):

        # Linear combination
        z = np.dot(X[i], weights) + bias
        y_pred = activation(z)

        error = y[i] - y_pred

        # Update rule
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        print(f"Items: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}, Error: {error}")

print("\nFinal Weights:", weights)
print("Final Bias:", bias)

# Step 5: Test the perceptron
test_items = np.array([2, 3, 6])

for items in test_items:
    z = np.dot(items, weights) + bias
    prediction = activation(z)

    if prediction == 1:
        result = "Large Order"
    else:
        result = "Small Order"

    print(f"Items Ordered: {items}, Prediction: {result}")
```
