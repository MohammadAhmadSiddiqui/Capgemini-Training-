```python
import pandas as pd

customers = pd.read_csv("Customers.csv")
sales = pd.read_csv("Sales.csv")
support = pd.read_csv("support.csv")

```

```python
print("Customers Info:")
print(customers.info())
print("Sales Info:")
print(sales.info())
print("Support Info:")
print(support.info())
```

```python
customers["Age"].fillna(customers["Age"].median(), inplace=True)
```

```python
print(customers.info())
```

```python
customers.rename(columns={"CustomerID": "Cust_ID"}, inplace=True)
sales.rename(columns={"CustomerID": "Cust_ID"}, inplace=True)
support.rename(columns={"CustomerID": "Cust_ID"}, inplace=True)

```

```python
sales["DiscountedPrice"]=sales["Price"]*0.9
sales["Revenue"]=sales["Quantity"]*sales["Price"]
```

```python
may_orders = sales[sales["OrderDate"].str.startswith("2023-05")]
print("May Orders:\n",may_orders)
print("First 10 Sales Records:\n", sales.head(10))
```

```python
north_customers = customers[customers["Region"] == "North"]
high_value_orders = sales[sales["Revenue"] > 10000]

print("\nNorth Customers:\n", north_customers)
print("\nHigh Value Orders:\n", high_value_orders)
```

```python
#Grouping
avg_revenue=sales.groupby("Product")["Revenue"].mean()
print(avg_revenue)
avg_resolution_issue=support.groupby("IssueType")["ResolutionTime"].mean()
print(avg_resolution_issue)
```

```python
#merge cutomers,sales,support datasets on customer_id
merged_data = sales.merge(customers, on="Cust_ID").merge(support, on="Cust_ID")
print("Merged Dataset:\n", merged_data.head())
```

```python
#Create new calculated fields:
#	Customer Lifetime Value (CLV) = total revenue per customer.
#	Average Resolution Time per Customer
clv = sales.groupby("Cust_ID")["Revenue"].sum().reset_index()
clv.rename(columns={"Revenue": "CustomerLifetimeValue"}, inplace=True)
customers = customers.merge(clv, on="Cust_ID", how="left")
print("Customers with CLV:\n", customers)
avg_resolution_time = support.groupby("Cust_ID")["ResolutionTime"].mean().reset_index()
avg_resolution_time.rename(columns={"ResolutionTime": "AverageResolutionTime"}, inplace=True)
customers = customers.merge(avg_resolution_time, on="Cust_ID", how="left")
print("Customers with Average Resolution Time:\n", customers)
```

```python
#export cleaned dataset
merged_data.to_csv("Cleaned_Data.csv", index=False)
print("Cleaned dataset exported to Cleaned_Data.csv")
```
