```python
#  Scenario Question
# A company wants to analyze whether advertising spend impacts sales revenue. The marketing team collected data from
#  several campaigns, where ad spend ranged from $2,000 to $18,000 and the corresponding sales revenue was recorded.
# You are asked to:
# - Plot a scatter chart showing the relationship between ad spend and sales.
# - Add a trend line to highlight the overall pattern.
# - Interpret the chart: does higher ad spend generally lead to higher sales?
#Ad spends vs Sales

ad_spend=np.array([2000,4000,6000,8000,10000,12000,14000,16000,18000])
sales=np.array([7000,15000,21000,25000,32000,36000,42000,47000,52000])

plt.scatter(ad_spend,sales,color='blue',alpha=0.7)
z=np.polyfit(ad_spend,sales,1)
p=np.poly1d(z)
plt.plot(ad_spend,p(ad_spend),"r--",label="Trend")
plt.xlabel("Ad Spend")
plt.ylabel("Sales")
plt.title("Ad Spend vs Sales")
plt.legend()
plt.show()
```

```python
import random
heads=0
total=1000
for i in range(total):
  result=random.randint(0,1)
  if result==1:
    heads=heads+1
tails=total-heads
print("Heads:",heads)
print("Tails:",tails)
print("P(Heads)=",heads/total)
print("P(Tails)=",tails/total)
```

```python
heights=np.random.normal(165,8,200)
print("Shortest",round(heights.min(),1))
print("Tallest",round(heights.max(),1))
print("Average",round(heights.mean(),1))
plt.hist(heights,bins=20,color='#00BFA6',edgecolor='white')
plt.axvline(heights.mean(),
            color='red',
            linewidth=2,
            label="Average")
plt.title("Height of 200 Students")
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
from scipy.stats import binom
n=10
p=0.25

p3=binom.pmf(3,n,p)
print("P(exactly 3) =",round(p3,3))
p5=1-binom.cdf(4,n,p)
print("P(5 or more) =",round(p5,3))

k=np.arange(0,11)
plt.bar(k,binom.pmf(k,n,p),
        color='#7B2EBE',
        edgecolor='white')
plt.title("Binomial Distribution")
plt.show()


```
